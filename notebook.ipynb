```python
x = df.iloc[:, [4, 7, 9, 10, 11]]
x
x
```

```python
y = df["Performance_Score"]
y
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
scaler = StandardScaler()
```

```python
scaled_x_train = scaler.fit_transform(x_train)
scaled_x_train
```

```python
import joblib
joblib.dump(scaler, "scaler.pkl")
```

```python
scaled_x_test = scaler.fit_transform(x_test)
scaled_x_test
```

```python
from sklearn.metrics import accuracy_score
```

```python
def model_performance(predict):
    print(f"Accuracy score on model is {accuracy_score(y_test, predict)}")
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression()
```

```python
lr.fit(scaled_x_train, y_train)
```

```python
lr.predict(scaled_x_test)
```

```python
y_pred = lr.predict(scaled_x_test)
```

```python
model_performance(y_pred)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
parameter_grid = {
    "n_neighbors" : [3,5,7, 11],
    "weights" : ["uniform", "distance"]
}
```

```python
gridkn = GridSearchCV(estimator=KNeighborsClassifier(), param_grid=parameter_grid, cv =5)
```

```python
gridkn.fit(scaled_x_train, y_train)
```

```python
gridkn.best_params_
```

```python
y_pred = gridkn.predict(scaled_x_test)
```

```python
model_performance(y_pred)
```

```python
from sklearn.svm import SVC
```

```python
s_model  = SVC()
```

```python
parameter_grid = {
    "C" : [0.01, 0.1, 1.0],
    "kernel" : ["linear","poly"]
}
```

```python
gridsvc = GridSearchCV(estimator=s_model, param_grid= parameter_grid, cv = 2)
```

```python
gridsvc.fit(scaled_x_train, y_train)
```

```python
y_pred = gridsvc.predict(scaled_x_test)
```

```python
model_performance(y_pred)
```
